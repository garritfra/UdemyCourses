```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset("iris")
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(sns.distplot, "total_bill")
```
